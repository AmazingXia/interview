```javascript
//  constructor 属性
// ✅ 定义：
// 所有 prototype 对象默认都有一个 constructor 指向它们的构造函数。

// ✅ 特点：
// 1. constructor 属性指向构造函数本身。
// 2. 可以通过实例的 __proto__ 访问到 constructor 属性。

// ✅ 作用：
// 保持构造函数和原型之间的联系；

// 可用于类型判断。


// Person.prototype.constructor === Person;
// ⚠️ 当你用 Object.create(...) 或手动修改 prototype 时，别忘了修复 constructor：



function Parent(name) {
  this.name = name
}

Parent.prototype.sayHello = function() {
  console.log(`Hello, my name is ${this.name}`);
};



function Child(name, age) {
  // 属性继承就一行代码   在子类中调用父类构造函数
  // 调用父类构造函数 相当于父类的初始化过程  在子类中重新执行一遍  这样子类中便拥有了父类所有定义的属性
  Parent.call(this, name);
  this.age = age;
}


// 要继承父类原型上的方法的话 需要两行代码
//  由于prototype上保存定义了共享的属性和方法 所以思路就是重写Child.prototype
// 直接定义 重写Child.prototype  但是由于重写后丢失constructor属性   需要重新赋值
Child.prototype = Object.create(Parent.prototype);
Child.prototype.constructor = Child;

Child.prototype.sayAge = function() {
  console.log(`I am ${this.age} years old.`);
};


```

```javascript
// 📌 1. 构造函数（Constructor）

// ✅ 定义：
// 构造函数是用于创建对象的函数，约定俗成首字母大写。

// ✅ 特点：
// 使用 new 调用时，会自动创建一个对象，并把 this 绑定到这个新对象。
// 默认返回这个新对象（如果没有手动 return 其他对象）。
// 创建的对象会自动指向该构造函数的 prototype 属性。




// 📌 2. 原型对象（Prototype）
// ✅ 定义：
// 每个函数都有一个 prototype 属性，这个属性是一个对象，用于存放将来所有实例共享的属性和方法。

Person.prototype.sayHi = function() {
  console.log('Hi, I am ' + this.name);
};

// ✅ 特点：
// 实例不会复制原型上的属性，而是通过原型链去访问。
// 原型对象上的属性是共享的，适合定义方法。
```
