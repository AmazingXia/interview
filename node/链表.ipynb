```javascript
// 单个节点类
class Node {
  constructor(element, next = null) {
    this.element = element
    this.next = next
  }
}

// 单向链表类
class LinkedList {
  constructor() {
    this.head = null
    this.size = 0
  }

  // 获取指定索引的节点
  _getNode(index) {
    if (index < 0 || index >= this.size) {
      throw new Error('越界了')
    }
    let currentNode = this.head
    for (let i = 0; i < index; i++) {
      currentNode = currentNode.next
    }
    return currentNode
  }

  // 添加节点
  add(index, element) {
    // 只传一个参数时，追加到末尾
    if (arguments.length === 1) {
      element = index
      index = this.size
    }
    if (index < 0 || index > this.size) {
      throw new Error('越界了')
    }

    if (index === 0) {
      this.head = new Node(element, this.head)
    } else {
      const prevNode = this._getNode(index - 1)
      prevNode.next = new Node(element, prevNode.next)
    }
    this.size++
  }

  // 删除指定索引的节点
  remove(index) {
    let rmNode = null
    if (index === 0) {
      rmNode = this.head
      if (!rmNode) return undefined
      this.head = rmNode.next
    } else {
      const prevNode = this._getNode(index - 1)
      rmNode = prevNode.next
      if (!rmNode) return undefined
      prevNode.next = rmNode.next
    }
    this.size--
    return rmNode.element
  }

  // 设置指定索引的值
  set(index, element) {
    const node = this._getNode(index)
    node.element = element
  }

  // 获取指定索引的值
  get(index) {
    return this._getNode(index).element
  }

  // 清空链表
  clear() {
    this.head = null
    this.size = 0
  }
}

// 基于链表的队列类
class Queue {
  constructor() {
    this.linkedList = new LinkedList()
  }

  enQueue(data) {
    this.linkedList.add(data)
  }

  deQueue() {
    return this.linkedList.remove(0)
  }
}

// 使用示例
const q = new Queue()
q.enQueue('node1')
q.enQueue('node2')
let a = q.deQueue()
console.log(a) // 输出：'node1'
```
