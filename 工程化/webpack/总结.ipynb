🧩 postcss-preset-env 插件详解
postcss-preset-env 是一个集合插件，它让你可以使用最新的 CSS 特性，同时自动向后兼容老版本浏览器。它类似于 Babel 的 @babel/preset-env。

| 插件名称                                 | 功能说明                                              |
| ------------------------------------ | ------------------------------------------------- |
| `autoprefixer`                       | 自动为 CSS 添加浏览器前缀（如 `-webkit-`, `-moz-`）            |
| `postcss-custom-properties`          | 支持 CSS 自定义属性（变量）                                  |
| `postcss-custom-media`               | 支持自定义媒体查询                                         |
| `postcss-media-minmax`               | 支持范围媒体查询（如 `width >= 500px`）                      |
| `postcss-nesting`                    | 支持 CSS 嵌套规则（类似 SCSS）                              |
| `postcss-color-functional-notation`  | 支持现代 CSS 颜色函数（如 `rgb(255 0 0 / 0.5)`）             |
| `postcss-color-mod-function`（已废弃）    | 曾用于 color-mod() 函数                                |
| `postcss-dir-pseudo-class`           | 支持 `:dir()` 伪类                                    |
| `postcss-logical`                    | 支持逻辑属性和方向（如 `margin-inline-start`）                |
| `postcss-lab-function`               | 支持 `lab()`、`lch()` 颜色函数                           |
| `postcss-focus-visible`              | polyfill `:focus-visible`                         |
| `postcss-focus-within`               | polyfill `:focus-within`                          |
| `postcss-attribute-case-insensitive` | 支持 `[type="button" i]` 语法                         |
| `postcss-env-function`               | 支持 `env()` 环境变量语法                                 |
| `postcss-gap-properties`             | 支持 `gap`、`row-gap`、`column-gap` 等属性               |
| `postcss-place`                      | 支持 `place-content`, `place-items`, `place-self` 等 |

## 10_webpack_importLoaders

问题:  css文件中使用了 @import语句导入了css文件, 但是postcss-loader并不认识@import语句




css-loader 配置项总结

| 选项                  | 类型                                             | 默认值       | 作用说明                                               |
| ------------------- | ---------------------------------------------- | --------- | -------------------------------------------------- |
| **`url`**           | `Boolean` \| `Function`                        | `true`    | 启用或禁用处理 `url()`，如 `background: url(image.png)`     |
| **`import`**        | `Boolean` \| `Function`                        | `true`    | 启用或禁用处理 `@import`                                  |
| **`modules`**       | `Boolean` \| `String` \| `Object`              | `false`   | 启用 CSS Modules（局部作用域）                              |
| **`sourceMap`**     | `Boolean`                                      | `false`   | 启用 source map                                      |
| **`importLoaders`** | `Number`                                       | `0`       | 设置在 `css-loader` 处理前，`@import` 引入的资源要经过的 loader 数量 |
| **`esModule`**      | `Boolean`                                      | `true`    | 是否以 ES Module 的方式导出（否则为 CommonJS）                  |
| **`exportType`**    | `'array'` \| `'string'` \| `'css-style-sheet'` | `'array'` | 控制导出 CSS 的类型（支持 Web Components）                    |

## webpack 4 与webpack5的区别





| 选项                | webpack4                                | webpack5                              | 作用说明                   |
|---------------------|-----------------------------------------|---------------------------------------|----------------------------|
| `require`           | `require('./111.png')`                  | `require('./111.png').default`        | img.src 导入项目图片        |
| `require` 返回值     | CommonJS 导出                           | 返回的是 ESModule                     | img.src 导入项目图片        |
| 处理图片资源          |    file-loader、url-loader、raw-loader      | asset module type                     | 处理图片资源                |
| `[ext]` 占位符       | 不带点（如 `png`）                      | 带点（如 `.png`）                     | 文件后缀                    |
| 文件资源             | file-loader                            | asset/resource        | 文件资源              |
| 文件资源             |  raw-loader                               | asset/source      | 文件资源              |
| 文件资源             |  url-loader                            | asset/inline   | 文件资源              |
